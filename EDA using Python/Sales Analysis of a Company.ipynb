## Visualising Sales month by month

```python
sns.barplot(x='month', y='Sales',data=df_train, palette='summer')
plt.xlabel('Months')
plt.ylabel('Sales')
```

## Inference:
1. During start of the years, the sale is comparatively higher.
2. The sale on month of August is more than it's neighbouring months. 

## Visualising sales Year by Year

```python
sns.barplot(x=df_train['year'], y=df_train['Sales'],data=df_train,palette='summer')
plt.xlabel('Years')
plt.ylabel('Sales')
```

## Inferences:
1. The sales have dropped in 2016 and 2018
2. The highest sale was on 2015 and lowest on 2018

```python
df_train['Category'].unique()
```

```python
df_cat=df_train.groupby(['Category']).sum().reset_index()
df_cat
```

## Visualising Sales according to Category

```python
plt.figure(figsize=(18,8))
colors = sns.color_palette('crest')[1:5]
plt.pie(x= df_cat['Sales'], labels=df_cat['Category'], explode=[0.1,0.1,0.1], colors=colors, autopct='%1.4f%%')
```

## Inferences:
1. Technology is the most bought category of products, followed by Furniture and Office Supplies.

## Top 5 cities according to Cities

```python
df_train['City'].unique()
```

```python
df_cit=df_train.groupby(['City']).sum().reset_index()
df_cit
```

```python
df_cit.sort_values(by=['Sales'], ascending=False, inplace=True)
df_cit.head(10)
```

```python
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
plt.figure(figsize=(18,8))
colors = sns.color_palette('viridis')[1:6]
plt.pie(df_cit['Sales'][:5], labels= df_cit['City'][:5],autopct='%1.4f%%', pctdistance=0.85,explode=explode,colors=colors)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
```

## Inferences
1. The sale in New York City is the highest, followed by Los Angeles.
2. The 3rd,4th and 5th city(ranked according to sales) are as follows: Seattle,San Fransisco,Philadelphia
